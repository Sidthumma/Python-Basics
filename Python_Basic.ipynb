1. What is Python, and why is it popular?
-> Python is a high-level, interpreted programming language known for its simplicity and readability. It supports multiple programming paradigms has a vast library ecosystem and is widely used in data science, web development, AI, and more. Its versatility and beginner friendly syntax make it a popular choice.

2. What is an interpreter in Python?
-> An interpreter in Python is a program that executes Python code line by line. It translates high-level Python code into machine-readable code, enabling immediate feedback and debugging. This approach makes Python suitable for rapid prototyping.

3. What are pre-defined keywords in Python?
-> Pre-defined keywords are reserved words in Python that have specific meanings and purposes in the language, such as if, else, while, and def. They cannot be used for variable or function names. These keywords form the syntax and structure of Python programs.

4. Can keywords be used as variable names?
-> No, keywords cannot be used as variable names in Python. Using them as identifiers will result in a syntax error since they are reserved for specific functionalities in the language.

5. What is mutability in Python?
-> Mutability refers to an object's ability to be modified after it is created. Mutable objects like lists and dictionaries can be changed, while immutable objects like strings and tuples cannot be altered after creation.

6. Why are lists mutable, but tuples are immutable?
-> Lists are mutable because their elements can be added, removed, or updated after creation. Tuples are immutable to ensure data integrity and allow them to be used as keys in dictionaries or stored in hash-based collections.

7. What is the difference between “==” and “is” operators in Python?
-> The == operator checks if two objects have the same value, while the is operator checks if they refer to the same memory location. For example, 5 == 5 is True, but list1 is list2 may be False if they are different objects.

8. What are logical operators in Python?
-> Logical operators in Python are used to combine conditional statements. The main operators are and both conditions must be true or at least one condition must be true and not reverses the condition.

9. What is type casting in Python?
-> Type casting in Python is the process of converting one data type to another, such as int to float or str to int. It can be achieved using functions like int(), float(), and str().

10. What is the difference between implicit and explicit type casting?
->  Implicit type casting is automatically performed by Python when mixing data types example int to float Explicit type casting requires the programmer to manually convert types using functions like int() or float().

11. What is the purpose of conditional statements in Python?
->  Conditional statements allow decision-making in Python programs. They execute specific blocks of code based on whether a condition evaluates to True or False, enabling dynamic program behavior.

12. How does the elif statement work?
->  The elif statement allows multiple conditions to be checked sequentially. If a condition is True, the corresponding block executes, and the rest of the conditions are ignored.

13. What is the difference between for and while loops?
->  A for loop iterates over a sequence e.g list, string for a specified number of times. A while loop continues execution as long as a given condition is True, making it suitable for indefinite iterations.

14. Describe a scenario where a while loop is more suitable than a for loop.
->  A while loop is more suitable when the number of iterations is not known in advance. For example, reading user input until they enter a specific value is better handled with a while loop.
